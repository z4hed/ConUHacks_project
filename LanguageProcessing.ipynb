```python
df = df.dropna(subset=['description'])
```

```python
import datetime
```

```python

def convertEpoch(epoch_time):
    datetime_time = datetime.datetime.fromtimestamp(epoch_time)
    return (datetime_time)
    
```

```python
new = []
for i in df['time']:
    new.append(convertEpoch(i))
    print(convertEpoch(i))

print(len(new))
```

```python
corpus = []
for i in  df['description']:
    description = re.sub("[^a-zA-Z]", ' ',i)
    description = description.lower()
    description = description.split()
    # Stemming which changes things like returned and loved to return and love respectively
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    ps = PorterStemmer()
    description = [ps.stem(word) for word in description if not word in set(all_stopwords)]
    description = ' '.join(description)
    corpus.append(description)
    
print(corpus)




```

## Calculate the sentiment

```python
print(len(corpus))
```

```python
df['description'] = corpus
df['description']
```

```python

df['polarity'] = df['description'].apply(lambda x: TextBlob(x).sentiment[0])
df['subjectivity'] = df['description'].apply(lambda x: TextBlob(x).sentiment[1])
```
